```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
from scipy.io.wavfile import write 
```

```python
mps=44100 #muestras por segundo
frec_portadora = 440 #hz
frec_moduladora = 0.25 #hz
duracion = 10 #hz
```

```python
muestras = np.arange(mps*duracion)
```

```python
portadora = np.sin(2*np.pi*frec_portadora* muestras/mps)
portadora *=0.3
portadora_ints = np.int16(portadora*32767)
```

```python
t1 = np.arange(0,10,0.01)# para senial portadora
```

```python
plt.plot(t1[:1000],portadora[:1000])
```

```python
moduladora = np.sin(2*np.pi*frec_moduladora*muestras/mps)
```

```python
graf_portadora=plt.plot(moduladora)
plt.xlabel("[t]s")
plt.ylabel("[amplitud]")
```

```python
m=12
modulante = portadora*np.sin(2*np.pi*frec_portadora*(muestras*mps)+m*np.cos(2*np.pi*frec_moduladora*(muestras/mps)))
```

```python
t3=np.arange(0.0,44100,0.1)# para la resultante
t3=np.radians(t3)
```

```python
plt.plot(modulante)
```

```python
write('señal-resultante-Fm.wav', mps, portadora_ints)
```
